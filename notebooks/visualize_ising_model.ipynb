```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.animation as animation
```

```python
plt.style.use('ggplot')
```

```python
file = pd.read_csv('../target/debug/states.csv', header=None)
lattice_size = 10
num_iters = 4000

select = [i for i in range(lattice_size)]
data = file[select]
```

```python
lattice = data[0:10]
lattice
```

```python
def plotanim(data):
    fig, ax = plt.subplots(figsize=(4, 4))

    def animate(value):
        ax.cla()
        idx = value * 10
        next_idx = (value + 1) * 10       
        d = data[idx:next_idx]
        ax.pcolor(d)
        ax.set_title(f'iter = {value}')
        return

    anim = animation.FuncAnimation(fig, animate, np.arange(500), interval=50, blit=False)
    
    return anim
```

```python
anim = plotanim(data)
anim.save('../figs/animation.gif', writer='imagemagick', fps=60)
```
